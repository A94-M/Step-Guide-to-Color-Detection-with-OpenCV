```python
import cv2
import pyttsx3
engine = pyttsx3.init()
engine.setProperty('rate',300) 
################################################
cap = cv2.VideoCapture(0)
cap.set(3,300)
cap.set(4,300)
################################################
while True:
    _, frame = cap.read()
    frame=cv2.resize(frame,(700,600))
    hsv_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    height, width, _ = frame.shape

    cx = int(width / 2 )
    cy = int(height / 2)

    # Pick pixel value
    pixel_center = hsv_frame[cy, cx]
    hue_value = pixel_center[0]
    #identify the color
    color = "Undefined"
    if hue_value < 5:
        color = "RED"
    elif hue_value < 22:
        color = "ORANGE"
    elif hue_value < 33:
        color = "YELLOW"
    elif hue_value < 78:
        color = "GREEN"
    elif hue_value < 131:
        color = "BLUE"
    elif hue_value < 170:
        color = "VIOLET"
    else:
        color = "Undefined"
    #pixeling the center 
    pixel_center_bgr = frame[cy, cx]
    b, g, r = int(pixel_center_bgr[0]), int(pixel_center_bgr[1]), int(pixel_center_bgr[2])
    #drawing the archtiecture
    cv2.rectangle(frame, (cx - 220, 40), (cx + 200, 120), (255, 255, 255), -1)
    cv2.putText(frame, color, (cx - 200, 100), 0, 2, (b, g, r), 5)
    cv2.circle(frame, (cx, cy), 12, (0, 250, 0), 2)
    cv2.imshow("Frame", frame) 
    
    
    #out.write(frame) #save your video
    key = cv2.waitKey(1)
    if key == 27:
        break
#out put
cap.release()

cv2.destroyAllWindows()
```
